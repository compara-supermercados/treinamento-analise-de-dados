```python
#Zerando as listas:
custo_produtos = []
venda_produtos = []
#Pegando o preço de custo de todos os produtos:
custo1 = int(input("Digite o valor de Custo do Produto 1: "))
custo2 = int(input("Digite o valor de Custo do Produto 2: "))
custo3 = int(input("Digite o valor de Custo do Produto 3: "))
custo4 = int(input("Digite o valor de Custo do Produto 4: "))
custo5 = int(input("Digite o valor de Custo do Produto 5: "))
#Pegando o preço de venda de todos os produtos:
venda1 = int(input("Digite o valor de Venda do Produto 1: "))
venda2 = int(input("Digite o valor de Venda do Produto 2: "))
venda3 = int(input("Digite o valor de Venda do Produto 3: "))
venda4 = int(input("Digite o valor de Venda do Produto 4: "))
venda5 = int(input("Digite o valor de Venda do Produto 5: "))
#Criando Array com os valores de custo digitados acima:
custo_produtos.append(custo1)
custo_produtos.append(custo2)
custo_produtos.append(custo3)
custo_produtos.append(custo4)
custo_produtos.append(custo5)
#Criando Array com os valores de venda digitados acima:
venda_produtos.append(venda1)
venda_produtos.append(venda2)
venda_produtos.append(venda3)
venda_produtos.append(venda4)
venda_produtos.append(venda5)
array_custo = np.array(custo_produtos)
array_venda = np.array(venda_produtos)
lucro = array_venda - array_custo
print(lucro)
```

```python
salarios = np.array([2,3,4,5,6,500])
maior_salario = np.max(salarios)
print(maior_salario)
```

```python
menor_salario = np.min(salarios)
print(menor_salario)
```

```python
media_salarios = np.mean(salarios)
print(f"{media_salarios:.2f}")
```

```python
mediana_salarios = np.median(salarios)
print(mediana_salarios)
```

```python
#Desvio Padrão
desvio = np.std(salarios)
print(desvio)
```

```python
#Criação de Array Vendas
vendas = np.array([10,20,45,5,90])
#Pegando a maior venda da Array Vendas
maior_venda = np.max(vendas)
#Pegando a menor venda da Array Vendas
menor_venda = np.min(vendas)
#Pegando a média de venda da Array Vendas
media_venda = np.mean(vendas)
#Pegando a mediana de venda da Array Vendas
mediana_venda = np.median(vendas)
#Exibindo a maior venda
print(f"A maior venda foi: {maior_venda}")
#Exibindo a menor venda
print(f"A menor venda foi: {menor_venda}")
#Exibindo a média das vendas
print(f"A média de Venda foi: {media_venda}")
#Exibindo a mediana de vendas
print(f"A mediana de Venda foi: {mediana_venda}")
```
