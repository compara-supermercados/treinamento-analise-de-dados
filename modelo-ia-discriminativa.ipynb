>Lendo as bibliotecas para usá-las

```python
import pandas as pd
import seaborn as sns
```

>Criando Gráfico

```python
df = pd.read_csv(r".\data\dados_tratados.csv")
sns.lmplot(data=df,x="study_hours_per_day",y="exam_score");
```

>Criando X e Y de acordo com os campos da tabela

```python
X = df[["study_hours_per_day","social_media_hours","netflix_hours","age","sleep_hours","attendance_percentage","exercise_frequency","mental_health_rating"]]
y = df["exam_score"]
```

>Treinando o modelo 

```python
from sklearn.linear_model import LinearRegression
modelo = LinearRegression()
modelo.fit(X,y)
```

>Criando DataFrame com os dados para fazer a consulta

```python
novo_aluno = pd.DataFrame({"study_hours_per_day":[4],"social_media_hours":[1],"netflix_hours":[1],"age":[20],"sleep_hours":[8],"attendance_percentage":[2],"exercise_frequency":[5],"mental_health_rating":[7]})
```

>Prevendo a nota de acordo com os dados acima

```python
resultado = modelo.predict(novo_aluno)
print(resultado)
```

>Mostra os coeficientes lineares dos campos selecionados da tabela (a porcentagem que altera a nota)

```python
#Coeficiente angular
modelo.coef_
```

>Mostra o coeficiente Linear da tabela

```python
#Coeficiente Linear
modelo.intercept_
```

>Mostra a porcentagem de confiança da nota de acordo com os dados selecionados da tabela

```python
modelo.score(X,y)
```
