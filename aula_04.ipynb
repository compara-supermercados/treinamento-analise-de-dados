```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv(r".\data\dataset_aula01_techshop_vendas.csv",parse_dates=["data_pedido"])
```

```python
grafico = sns.countplot(data=df,x="categoria");
grafico.set_title("Quantidade de Produtos por Categoria");
grafico.set_ylabel("Quantidade");
grafico.set_xlabel("Categorias");
```

```python
grafico = sns.histplot(data=df,x="preco_unitario",kde=True,bins=100);
grafico.set_title("Histograma");
grafico.set_ylabel("Quantidade de Vendas")
grafico.set_xlabel("Preços dos Produtos")
```

```python
grafico = sns.boxplot(data=df,x="categoria",y="preco_unitario")
grafico.set_title("Distribuição de Preços por Categoria")
grafico.set_xlabel("Categorias")
grafico.set_ylabel("Preços")
```
