```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error

# Selecionar as variáveis numéricas que fazem sentido para a rotina habitual
features = ['study_hours_per_day', 'social_media_hours', 'netflix_hours', 
            'sleep_hours', 'exercise_frequency', 'mental_health_rating', 'attendance_percentage']
X = df[features]
y = df['exam_score']

# Treinar um modelo simples de Regressão Linear
model = LinearRegression()
model.fit(X, y)

# Mostrar os coeficientes para entender o peso real de cada um no modelo
coefficients = pd.DataFrame({'Hábito': features, 'Impacto na Nota (Pontos)': model.coef_})
print(coefficients.sort_values(by='Impacto na Nota (Pontos)', ascending=False))

# Salvar o modelo e as features no escopo global para usar depois se necessário
def prever_nota(estudo, redes, netflix, sono, exercicio, saude_mental, presenca):
    entrada = [[estudo, redes, netflix, sono, exercicio, saude_mental, presenca]]
    nota = model.predict(entrada)[0]
    return min(max(nota, 0), 100) # Garantir que a nota fique entre 0 e 100

print("\nTeste de previsão (Rotina Equilibrada):", prever_nota(4, 1.5, 1, 7, 3, 7, 90))
print("Teste de previsão (Rotina Desfocada):", prever_nota(1, 4, 3, 6, 0, 4, 80))

```

### 1. Horas de estudo → impacto massivo

## Parte 3

Análise: Por que os alunos mais velhos parecem tirar notas mais altas?
📊 O que os dados mostram
Primeiro, o resultado mais surpreendente: a correlação geral entre idade e nota é praticamente zero (−0,009). Ou seja, olhando a base inteira, idade não explica o desempenho. No entanto, quando agrupamos em faixas etárias, a percepção muda:


Faixa etária	Nota média	Nota mediana	Contagem
17-18	70,6	70,9	250
19-20	69,2	70,6	262
21-22	68,3	69,0	238
23-24	70,2	70,7	253
De fato, os mais jovens (17-18) e os mais velhos (23-24) têm notas médias mais altas que o grupo intermediário (19-22). Mas o teste t entre os extremos (17-18 vs 23-24) mostra que essa diferença não é estatisticamente significativa (p = 0,798).



---

🔍 As 3 razões para essa percepção
1. Efeito de composição (confusão) — a principal explicação
A variável que mais se correlaciona com a nota é a escolaridade dos pais, que também se distribui de forma diferente entre as faixas etárias:

Nos 23-24 anos, 38% têm pais com diploma de Bachelor (a escolaridade que gera as notas mais altas: 70,3)
Nos 17-18 anos, apenas 34% têm pais com Bachelor, mas 18% têm pais com Master (que, curiosamente, gera as menores notas: 68,1) 

Quando olhamos a nota dentro de cada nível educacional dos pais, a idade praticamente não faz diferença:

Nível dos pais	17-18	19-20	21-22	23-24
Bachelor	71,1	70,4	69,5	70,1
High School	70,0	70,1	69,0	69,0
Master	69,2	67,1	66,1	69,7
➡️ Conclusão: a diferença que você percebe não é causada pela idade, mas pela composição diferente de perfis em cada faixa etária.

2. Distribuição desigual de gênero

Os mais velhos (23-24) têm mais homens (53%) e menos mulheres (42%), enquanto nos mais jovens a proporção é mais equilibrada (~49% homens). Como homens tendem a ter notas ligeiramente diferentes de mulheres na base, isso também contribui para a percepção de diferença entre faixas.

3. Variação amostral natural (ruído)

A diferença entre as médias das faixas é pequena (~2 pontos) e o R² da regressão idade+educação dos pais explica apenas ~1,3% da variação da nota. Isso é consistente com a hipótese de que a variação observada é ruído amostral, não um padrão real.

✅ Resumo direto

Os alunos mais velhos não tiram notas genuinamente mais altas. A impressão vem de um efeito de composição: os grupos de idade têm perfis diferentes quanto à escolaridade dos pais (a variável que mais impacta a nota), gênero e outras características. Quando você controla por esses fatores, a idade perde completamente o efeito — a correlação parcial é −0,009, essencialmente zero.

A "diferença" que você observou é um exemplo clássico de paradoxo de Simpson: um padrão que aparece em grupos agregados, mas desaparece quando controlamos pelas variáveis de confusão reais.
