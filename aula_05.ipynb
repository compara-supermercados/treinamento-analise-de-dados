```python
#Separando campos com números para gerar gráficos
colunas = ["age","study_hours_per_day","social_media_hours","netflix_hours","attendance_percentage","sleep_hours","exercise_frequency","mental_health_rating","exam_score"]
```

>Cria uma variável **colunas** somente com as colunas do tipo numéricas (int ou float), excluindo as variáveis str (string) e outras do tipo somente texto.

```python
sns.heatmap(data=df[colunas].corr(),annot=True,fmt=".2f",cmap="coolwarm");
```

**RESULTADOS DO GRÁFICO ACIMA:**

>Quanto mais próximo do 1, maior será a relação e quanto mais próximo do -1 menor será a relação. E o 0 (zero) a relação é nula.

1 - A idade dos participantes do estudo é irrelevante para a nota do aluno.
2 - A quantidade de horas de estudo por dia influencia positivamente na nota do aluno.
3 - A quantidade de tempo em redes sociais influenciam negativamente na nota do aluno.
4 - A quantidade de horas de Netflix influenciam negativamente na nota do aluno.
5 - Quanto mais o aluno frequentar as aulas melhor a nota do aluno.
6 - Quanto mais o aluno dormir, maior será a nota dele.
7 - Quanto mais exercício físico o aluno fizer, maior será a nota dele.
8 - Quanto maior for a saúde mental do aluno, maior será a nota dele.

```python
filtro_horas_altas = df["study_hours_per_day"] >= 5

filtro_horas_baixas = df["study_hours_per_day"] < 3
```

>Gera as variáveis **filtro_horas_altas** notas dos alunos que estudaram 5 ou mais horas por dia e a variável **filtro_horas_baixas** nota dos alunos que estudaram menos que 3 horas.

```python
media_pessoas_estudam_muito = df[filtro_horas_altas]["exam_score"].mean()

media_pessoas_estudam_pouco = df[filtro_horas_baixas]["exam_score"].mean()

print(f"A média das notas de pessoas que estudam 5 ou mais horas por dia é de: {media_pessoas_estudam_muito:,.2f}")

print(f"A média das notas de pessoas que estudam menos de 3 horas por dia é de: {media_pessoas_estudam_pouco:,.2f}")
```

>Mostra a média dos alunos conforme a quantidade de horas de estudo.

```python
sns.lmplot(data=df,x="study_hours_per_day",y="exam_score");
```

>Cria gráfico de regressão e de dispersão. Esse gráfico mostra a relação de quanto mais horas por dia de estudo maior a nota do aluno.

```python
sns.lmplot(data=df,x="netflix_hours",y="exam_score");
```

>Acima o gráfico mostra que quanto mais tempo assistindo Netflix, menor a nota do aluno.

```python
sns.lmplot(data=df,x="age",y="exam_score");
```
