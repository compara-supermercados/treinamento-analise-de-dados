Conceito de n-gramas

```python
frase = "O amor é lindo o amor é forte e o amor vence tudo"
```

```python
palavras = frase.split()
palavras
```

```python
import pandas as pd
tabela = pd.DataFrame({"palavra":palavras})
tabela
```

```python
tabela["proxima_palavra"] = tabela["palavra"].shift(-1)
tabela
```

```python
tabela = tabela.dropna()
tabela
```

```python
filtro = tabela["palavra"] == "lindo"
tabela[filtro]
```

```python
tabela[filtro]["proxima_palavra"].value_counts(normalize=True)
```

```python
tabela[filtro]["proxima_palavra"].value_counts().idxmax()
```
